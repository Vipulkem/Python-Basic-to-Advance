# multithreading :-
- Multithreading in python allows a program to run multiple folws of exection(thread) concurrently within a single process

```python
import threading
import time
```

```python
def square(num):
    print(f"Square of {num} is {num * num}")
    time.sleep(1)

def cube(num):
    print(f"Cube of {num} is {num * num * num}")
    time.sleep(1)

t1 = threading.Thread(target=square, args=(4,))
t2 = threading.Thread(target=cube, args=(4,))

t1.start()
t2.start()
t1.join()
t2.join()

print("Done!")
```

```python
from concurrent.futures import ThreadPoolExecutor

def worker(task):
    time.sleep(2)
    print(f"Processing {task}")
    time.sleep(5)
    return f"Result of {task}"


with ThreadPoolExecutor(max_workers=2) as executor:
    tasks = ['Task_1', 'Task_2', 'Task_3', 'Task_4']
    results = list(executor.map(worker, tasks))
print(results)

```
