# Loops:- 
- Loops in Python are programming constructs used to repeat a specific block of code multiple times, which helps automate repetitive tasks and reduce code redundancy. Python primarily supports two types of loops: for loops and while loops.

1. The for Loop
- A for loop is used to iterate over a sequence (such as a list, tuple, string, or range) or any other iterable object. It executes a block of code once for each item in the sequence.

```python
for i in range(5):
    print(i)
```

```python
for i in range(5, 10):
    print(i)
```

- Best Used When: You know in advance how many times the code needs to run or you are processing a collection with a fixed set of items.

```python
lst = ['banana', 'apple', 'grape',23, 45.6, True, None]
for item in lst:
    print(item)
```

```python
for i in 'banana':
    print(i)
```

2. The while Loop
- A while loop repeatedly executes a block of code as long as a specified Boolean condition remains True. 

- Best Used When: The number of iterations is not known beforehand and depends on a dynamic condition, such as waiting for user input or reaching a calculation threshold.

```python
i = 0
while True:
    print("This will run forever")
# break statement
    if i == 5:
        break
    i += 1
```

3. Loop Control Statements
- These keywords allow you to change the normal flow of a loop's execution: 

- i.  break: Immediately exits the loop entirely.
- ii. continue: Skips the current iteration and jumps to the next one.
- iii. pass: A null operation used as a placeholder when a statement is required syntactically but no action is needed yet.

```python
# break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue statement
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass statement
for i in range(10):
    if i % 2 == 0:
        pass
    else:
        print(i)
```

4. Advanced Concepts
- Nested Loops: You can place one loop inside another. The inner loop executes all its iterations for every single iteration of the outer loop.
- Loop else Clause: Python allows an else block at the end of a loop. It runs only if the loop finishes normally (i.e., it wasn't stopped by a break statement).
- List Comprehensions: A more concise, "Pythonic" way to create lists by applying an expression to each item in an iterable, often replacing simple for loops.

```python
# neasted Loop
for i in range(3):
    for j in range(3):
        print(i, j)
```

```python
# Loops else statement
for i in range(5):
    print(i)
else:
    print("Loop is completed")
```

```python
# List comprehension
squares = [x**2 for x in range(10)]
print(squares)
```
