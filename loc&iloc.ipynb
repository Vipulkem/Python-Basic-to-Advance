# loc:-
- In pandas, loc is a property used to access and manipulate data in a DataFrame or Series based on labels (names of rows and columns) rather than integer positions.

## Key Features
- Label-Based: You use actual row and column names (e.g., 'Alice', 'City') to retrieve data.
- Inclusive Slicing: Unlike standard Python slicing, loc includes both the start and end labels (e.g., df.loc['A':'C'] includes row 'C').
- Boolean Support: It can filter rows based on conditions (e.g., df.loc[df['Age'] > 30]).
- Updating Data: It is the preferred method for modifying specific values in a DataFrame to avoid "Chained Assignment" warnings.

## Basic Syntax
- The standard format for the pandas.DataFrame.loc property is:
df.loc[row_label, column_label]

## Usage Examples
- Single Value: df.loc['row1', 'col1'] returns the value at that specific intersection.
- Full Row: df.loc['row1'] or df.loc['row1', :] returns all columns for a specific row.
- Multiple Rows/Columns: df.loc[['A', 'B'], ['Name', 'Age']] uses lists to select specific subsets.
- Slicing: df.loc['row1':'row5', 'colA':'colC'] selects a range of data inclusive of endpoints.
- Filtering: df.loc[df['Salary'] > 50000] returns rows where the condition is met

# iloc

- In Pandas, iloc is a property used for integer-location based indexing, allowing you to select data from a DataFrame or Series by its numerical position rather than its label.

## Key Characteristics
- Zero-based: Indexing starts at 0 for both rows and columns.
- Exclusive Endpoints: When using slices (e.g., 0:5), the end index is excluded (returns indices 0, 1, 2, 3, and 4).
- Strict Integers: It primarily accepts integers, lists of integers, or slice objects

## Common Syntax & Usage
- The general syntax is df.iloc[row_indexer, column_indexer].

- Select a single row: df.iloc[0] retrieves the first row as a Series.
- Select a single value: df.iloc[1, 2] retrieves the value at the second row and third column.
- Slicing rows and columns: df.iloc[0:3, 1:4] selects rows 0-2 and columns 1-3.
- Selecting specific subsets: df.iloc[[0, 2], [0, 1]] selects the first and third rows for the first two columns.
- Select all rows for specific columns: df.iloc[:, [0, 2]] retrieves every row for the first and third columns

```python
# Select a single row: 
df.iloc[0]
```

```python
# select a single value by row and column index:
df.iloc[4, 6] # [row index, column index]
```

```python
# Slicing rows and columns:
df.iloc[0:5, 0:4] # [row start:row end, column start:column end]
```

```python
# Selecting specific subsets:
df.iloc[[0, 2, 4], [1, 3]] # [row indices, column indices]
```

```python
# Select all rows for specific columns: 
df.iloc[:, [1, 3]] # [all rows, column indices]
```

## difference between loc and iloc:
- The primary difference between loc and iloc in Pandas is that loc is label-based (using row/column names), while iloc is integer-position based (using 0-based indices). loc includes the last element in slices, whereas iloc excludes it. loc supports boolean indexing; iloc does not.
