```python
for i in range(5):
    print("*"*i)
```

```python
n = 5
for i in range(n):
    print(" " * (n - i - 1) + "*" * (2 * i + 1))
```

```python
# Inverted Pyramid: The reverse of a pyramid, starting with the maximum stars and decreasing.
n = 5
for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
n = 5
# Upper part
for i in range(n):
    print(" " * (n - i - 1) + "*" * (2 * i + 1))
# Lower part
for i in range(n - 2, -1, -1):
    print(" " * (n - i - 1) + "*" * (2 * i + 1))
```

```python
# Hollow Square Pattern: Uses conditional logic to print stars only on the borders.
n = 5
for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
