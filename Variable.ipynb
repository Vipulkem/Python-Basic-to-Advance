# Variable :-
- A variable in Python is a named container or reference used to store data values in a computer's memory. Unlike languages like Java, Python does not require you to declare a variable's type; it is created automatically the moment you assign it a value using the equals sign (=).

```python
a = 25
print(a)
```

# Key Characteristics :- 
- Dynamic Typing: Python automatically determines the data type (e.g., integer, string) based on the assigned value. A variable can also change types if you assign it a new kind of value later.
- Object References: Technically, Python variables are pointers to objects stored in memory rather than the actual memory location itself.
- Case-Sensitivity: Variable names are case-sensitive. For example, Age, age, and AGE are treated as three distinct variables.

# Rules for Naming Variables :-
- To create a valid variable name, you must follow these rules: 

- Start: Must begin with a letter (a-z, A-Z) or an underscore (_).
- Content: Can only contain alphanumeric characters and underscores (A-z, 0-9, and _).
- Forbidden: Cannot start with a number and cannot be a Python keyword like if, else, or while.
- Convention: The standard practice is to use snake_case (all lowercase with underscores between words), such as user_age or total_score.

```python
name = "John"
print(name)
```

# Common Data Types :- 
- Variables can hold various types of information, including: 
- Integers (int): Whole numbers like 5 or -10.
- Floating-point (float): Decimal numbers like 19.99.
- Strings (str): Text surrounded by quotes, like "Hello".
- Booleans (bool): Logical values, either True or False.
- Collections: Complex types like Lists [1, 2, 3], Tuples (1, 2), and Dictionaries {"key": "value"}. 

# Type and Casting a Variable:-
- We can determine the type of a variable using the type() function. This built-in function returns the type of the object passed to it.
- Type casting refers to the process of converting the value of one data type into another. Python provides several built-in functions to facilitate casting, including int(), float() and str() among others. For example, int() converts compatible values to an integer, float() to a floating point number, and str() to a string,

# Deleting a Variable :-
- We can remove a variable from the namespace using the del keyword. This deletes the variable and frees up the memory it was using.

```python
x = 10
del x
print(x)
```

- del x removes the variable x from memory.
- After deletion, trying to access the variable x results in a NameError indicating that the variable no longer exists.

# Practical Examples

# 1. Swapping Two Variables:-
- Using multiple assignments, we can swap the values of two variables without needing a temporary variable.

```python
a, b = 5, 10
a, b = b, a
print("a :", a, "b :", b)
```
