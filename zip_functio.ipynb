# Zip :-
- The term zip in Python typically refers to the built-in zip() function used for combining iterables, but it can also refer to managing compressed file archives using the zipfile module. 

```python
name = ["Rocky","Abhay","Vijay","Rohit","Satyarth"]
Roll_no = ["001","002","003","004","005"]

zipped = list(zip(name,Roll_no))
print(zipped)
```

- Parallel Iteration: It is most commonly used to loop through multiple lists simultaneously.
- Handling Unequal Lengths: By default, zip() stops when the shortest iterable is exhausted.
- Strict Mode: Since Python 3.10, you can set strict=True to raise a ValueError if lengths differ.
- Padding: Use itertools.zip_longest() to fill missing values with a default (like None) instead of stopping.

# unzipping :- 
-You can reverse a zip operation by using the unpacking operator *.

```python
zipped
```

```python
new_name,new_roll_no = zip(*zipped)
print(new_name)
print(new_roll_no)
```
