- In Python, a lambda function is a small, anonymous function defined without a name. Unlike standard functions defined with the def keyword, lambda functions are expressed in a single line and are typically used for short-lived tasks where a full function definition would be overkill. 

```python
a = lambda x : x + 1
print(a(5))
```

```python
fun = lambda x,y,z,c : x + y + z + c
print(fun(1,2,3,4))
```

```python
def square(x):
    return lambda x : x * x
```

```python
square(79)
```
