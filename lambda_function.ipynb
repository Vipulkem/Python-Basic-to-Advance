- In Python, a lambda function is a small, anonymous function defined without a name. Unlike standard functions defined with the def keyword, lambda functions are expressed in a single line and are typically used for short-lived tasks where a full function definition would be overkill. 

```python
a = lambda x : x + 1
print(a(5))
```

```python
fun = lambda x,y,z,c : x + y + z + c
print(fun(1,2,3,4))
```

```python
def square(x):
    return lambda x : x * x
```

```python
squ = square(79)
squ(3)
```

```python
for i in range(5,37,7):
    print(f'square of {i} : {squ(i)}')
```

```python
l = [2, 3, 4, 5, 6, 8, 45,23, 34, 90]

list(map(lambda x : 'even' if x%2==0 else 'odd', l))
```

```python
l = [2, 3, 4, 5, 6, 8, 45,23, 34, 90]

list(filter(lambda x : x%2 ==0, l))
```

```python
from functools import reduce
reduce(lambda x, y : x+y, [2, 3, 4, 5])   # 2+3+4+5 = 14
```
