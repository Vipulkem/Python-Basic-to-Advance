```python
l = [5, 2, 8, 1, 9, 3,3445,43545,432543,545,432,435,43546,546,423]
print("Original list:", l)
```

```python
for i in range(len(l)):
    for j in range(len(l) - i - 1):
        if l[j] > l[j + 1]:
            l[j], l[j + 1] = l[j + 1], l[j]

print("List after sorting:", l)
print("Second highest number:", l[-2])
```

```python
l = [5, 2, 8, 1, 9, 3,3445,43545,432543,545,432,435,43546,546,423]
print("Original list:", l)
l.sort()
print("List after sorting:", l)
print("Second highest number:", l[-2])
```
