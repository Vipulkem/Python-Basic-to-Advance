# List :- 
-> In Python, a list is a built-in data structure used to store a collection of items in a specific order. Lists are mutable, meaning you can change, add, or remove items after the list is created.
- list is used to store multiple values
- list is mutable(can Change).
- it is ordered
- it allows duplicate value.
- it is defined by [].

```python
# Create list :-
list = [1, 2, 3, 4, 5]
list_1 = ["apple", "banana", "cherry"]
list_2 = [1, "apple", 3.14, True]
list_3 = [1, 2, [3, 4], 5]
list_4 = []
print(list)
print(list_1)
print(list_2)
print(list_3)
print(list_4)
```

```python
# Access list items :-
my_list = ["apple", "banana", "cherry"]
print(my_list[0])  # Output: apple
print(my_list[1])  # Output: banana
print(my_list[2])  # Output: cherry
```

```python
# Length of list :-
my_list = ["apple", "banana", "cherry", True, 3.14, 9, [1, 2, 3]]
print(len(my_list))  # Output: 7
```

```python
# slicing list :-
my_list = ["apple", "banana", "cherry", "date", "fig", "grape"]
print(my_list[1:4])  # Output: ['banana', 'cherry', 'date']
print(my_list[:3])   # Output: ['apple', 'banana', 'cherry']
print(my_list[3:])   # Output: ['date', 'fig', 'grape']
print(my_list[-3:])  # Output: ['date', 'fig', 'grape']
print(my_list[::2])  # Output: ['apple', 'cherry', 'fig']   
```

```python
# Modify list :-
my_list = ["apple", "banana", "cherry"]
my_list[1] = "blueberry"
print(my_list)  # Output: ['apple', 'blueberry', 'cherry']
```

```python
# Adding items to list :-
my_list = ["apple", "banana", "cherry"]
my_list.append("date")
print(my_list)  # Output: ['apple', 'banana', 'cherry', 'date']
my_list.insert(1, "blueberry")  
print(my_list)  # Output: ['apple', 'blueberry', 'banana', 'cherry', 'date']
my_list.extend(["fig", "grape"])
print(my_list)  # Output: ['apple', 'blueberry', 'banana', 'cherry', 'date', 'fig', 'grape']
```

```python
# Remioving items from list :-
my_list = ["apple", "banana", "cherry", "date", "fig", "grape"]
my_list.remove("date")
print(my_list)  # Output: ['apple', 'banana', 'cherry', 'fig', 'grape'] 
my_list.pop(2)
print(my_list)  # Output: ['apple', 'banana', 'fig', 'grape']
my_list.clear()
print(my_list)  # Output: []
```

```python
my_list = ["apple", "banana", "cherry", "date", "fig", "grape"]
del my_list[1]
print(my_list)  # Output: ['apple', 'cherry', 'date', 'fig', 'grape']
del my_list[1:4]    
print(my_list)  # Output: ['apple', 'grape']
# del my_list
print(my_list)  # This will raise an error since my_list is deleted
```

```python
# Looping in List
vegitables = ["carrot", "broccoli", "spinach"]
for veg in vegitables:
    print(veg)
```

```python
for i in range(len(vegitables)):
    print(vegitables[i])
```

```python
# checking if item exists in list
my_list = ["apple", "banana", "cherry"]
for item in my_list:
    if item == "banana":
        print("Found banana!")
        break
```

```python
#List comprehension
squares = [x**2 for x in range(1,11)]
print(squares)
```
