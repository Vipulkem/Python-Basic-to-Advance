```python
import pandas as pd
```

```python
# Step 1: Create sample data (you can skip this if reading from file)
data = [
    ["A", "B", "C"],        # Row 1
    ["D", "E", "F"],        # Row 2
    ["Name", "Age", "City"],  # Row 3 (this should become header)
    ["Vipul", 22, "Delhi"],
    ["Rahul", 25, "Mumbai"]
]
```

```python
# Step 2: Convert to DataFrame
df = pd.DataFrame(data)
```

```python
print("Original DataFrame:\n")
print(df)
```

```python
# Step 3: Set 3rd row as header
df.columns = df.iloc[2]
```

```python

# Step 4: Remove first 3 rows
df = df[3:].reset_index(drop=True)
```

```python
print("\nUpdated DataFrame:\n")
print(df)
```

```python
df.reset_index(drop=True, inplace=True)
```
